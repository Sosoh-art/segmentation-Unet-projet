```python
# ========== IMPORTATION DES BIBLIOTHÈQUES ==========

# TensorFlow : Framework principal pour créer et entraîner des réseaux de neurones
import tensorflow as tf

# TensorFlow Datasets : Pour charger facilement des datasets standardisés
import tensorflow_datasets as tfds

# NumPy : Pour les opérations mathématiques sur des tableaux et matrices
import numpy as np

# Matplotlib.pyplot : Pour créer des graphiques et afficher des images
import matplotlib.pyplot as plt

# Keras layers : Contient les couches du réseau (Conv2D, MaxPooling, etc.)
from tensorflow.keras import layers, models

# Callbacks : Fonctions appelées pendant l'entraînement pour sauvegarder/arrêter
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping

# OpenCV : Bibliothèque de traitement d'images (lecture, redimensionnement, etc.)
import cv2


# ========== CONFIGURATION DE L'ALÉATOIRE ==========
# Pour avoir des résultats reproductibles à chaque exécution

# Fixe la graine aléatoire de TensorFlow à 42
# Cela garantit que les poids initiaux du réseau seront toujours les mêmes
tf.random.set_seed(42)

# Fixe la graine aléatoire de NumPy à 42
# Cela garantit que les opérations aléatoires (shuffle, etc.) seront identiques
np.random.seed(42)


# ========== VÉRIFICATION DU GPU ==========
# Vérifie si un GPU est disponible pour accélérer les calculs

# list_physical_devices('GPU') : Liste tous les GPU disponibles
# Si la liste est vide [], aucun GPU n'est détecté
# Si un GPU est présent, l'entraînement sera beaucoup plus rapide
print("GPU disponible:", tf.config.list_physical_devices('GPU'))
import tensorflow_datasets as tfds
dataset, info = tfds.load('oxford_iiit_pet', with_info=True)
```

```python
# ========== DÉFINITION DES PARAMÈTRES ==========

# IMG_SIZE : Taille à laquelle toutes les images seront redimensionnées
# 128x128 pixels est un bon compromis entre qualité et vitesse
IMG_SIZE = 128

# BATCH_SIZE : Nombre d'images traitées simultanément
# 16 images par batch est optimal pour la mémoire GPU
BATCH_SIZE = 16

# BUFFER_SIZE : Taille du buffer pour mélanger les données
# 1000 signifie que 1000 images seront mélangées aléatoirement
BUFFER_SIZE = 1000


# ========== FONCTION DE NORMALISATION ==========
def normalize(input_image, input_mask):
    """
    Normalise l'image et le masque pour améliorer l'entraînement

    Args:
        input_image : Image d'entrée (valeurs entre 0 et 255)
        input_mask : Masque de segmentation

    Returns:
        input_image : Image normalisée (valeurs entre 0 et 1)
        input_mask : Masque en float32
    """
    # tf.cast : Convertit le type de données
    # float32 : Type de données à virgule flottante (nombres décimaux)
    input_image = tf.cast(input_image, tf.float32)

    # Division par 255.0 : Normalise les valeurs de [0, 255] vers [0, 1]
    # Les réseaux de neurones apprennent mieux avec des valeurs entre 0 et 1
    input_image = input_image / 255.0

    # Convertit le masque en float32 pour les calculs
    input_mask = tf.cast(input_mask, tf.float32)

    # Retourne l'image et le masque normalisés
    return input_image, input_mask


# ========== FONCTION DE REDIMENSIONNEMENT ==========
def resize(input_image, input_mask):
    """
    Redimensionne l'image et le masque à la taille souhaitée

    Args:
        input_image : Image d'entrée (taille variable)
        input_mask : Masque de segmentation (taille variable)

    Returns:
        input_image : Image redimensionnée (IMG_SIZE x IMG_SIZE)
        input_mask : Masque redimensionné (IMG_SIZE x IMG_SIZE)
    """
    # tf.image.resize : Redimensionne une image à la taille spécifiée
    # [IMG_SIZE, IMG_SIZE] : Nouvelle taille [hauteur, largeur]
    # Par défaut, utilise l'interpolation bilinéaire (lissage des pixels)
    input_image = tf.image.resize(input_image, [IMG_SIZE, IMG_SIZE])

    # Redimensionne le masque de la même façon
    input_mask = tf.image.resize(input_mask, [IMG_SIZE, IMG_SIZE])

    # Retourne l'image et le masque redimensionnés
    return input_image, input_mask


# ========== FONCTION D'AUGMENTATION DES DONNÉES ==========
def augment(input_image, input_mask):
    """
    Applique des transformations aléatoires pour augmenter la diversité des données
    L'augmentation aide le modèle à mieux généraliser et éviter le surapprentissage

    Args:
        input_image : Image d'entrée
        input_mask : Masque de segmentation

    Returns:
        input_image : Image transformée
        input_mask : Masque transformé (avec les mêmes transformations)
    """

    # === FLIP HORIZONTAL (Miroir gauche-droite) ===
    # tf.random.uniform(()) : Génère un nombre aléatoire entre 0 et 1
    # > 0.5 : Vrai dans 50% des cas
    if tf.random.uniform(()) > 0.5:
        # flip_left_right : Inverse l'image horizontalement (effet miroir)
        # Important : On applique la même transformation à l'image ET au masque
        input_image = tf.image.flip_left_right(input_image)
        input_mask = tf.image.flip_left_right(input_mask)

    # === AJUSTEMENT DE LA LUMINOSITÉ ===
    # random_brightness : Change aléatoirement la luminosité de l'image
    # max_delta=0.2 : La luminosité peut varier de ±20%
    # Aide le modèle à gérer différentes conditions d'éclairage
    input_image = tf.image.random_brightness(input_image, max_delta=0.2)

    # === AJUSTEMENT DU CONTRASTE ===
    # random_contrast : Change aléatoirement le contraste de l'image
    # lower=0.8, upper=1.2 : Le contraste peut varier entre 80% et 120%
    # Aide le modèle à s'adapter à différentes qualités d'image
    input_image = tf.image.random_contrast(input_image, lower=0.8, upper=1.2)

    # Note : On ne modifie PAS le masque avec la luminosité/contraste
    # car ces transformations ne changent pas la position des objets

    # Retourne l'image et le masque augmentés
    return input_image, input_mask


# ========== FONCTION DE PRÉTRAITEMENT COMPLÈTE ==========
def preprocess(data, augment_data=False):
    """
    Pipeline complet de prétraitement des données
    Combine toutes les étapes : redimensionnement, augmentation, normalisation

    Args:
        data : Dictionnaire contenant l'image et le masque
        augment_data : Boolean, True pour appliquer l'augmentation

    Returns:
        input_image : Image prétraitée
        input_mask : Masque prétraité
    """
    # Extraction de l'image depuis le dictionnaire de données
    # data['image'] : Clé pour accéder à l'image dans le dataset
    input_image = data['image']

    # Extraction du masque de segmentation
    # data['segmentation_mask'] : Clé pour accéder au masque dans le dataset
    input_mask = data['segmentation_mask']

    # === ÉTAPE 1 : REDIMENSIONNEMENT ===
    # Appelle la fonction resize définie plus haut
    input_image, input_mask = resize(input_image, input_mask)

    # === ÉTAPE 2 : AUGMENTATION (OPTIONNELLE) ===
    # if augment_data : Vérifie si on doit augmenter les données
    # True pour l'entraînement, False pour la validation/test
    if augment_data:
        # Applique les transformations aléatoires
        input_image, input_mask = augment(input_image, input_mask)

    # === ÉTAPE 3 : NORMALISATION ===
    # Normalise les valeurs de pixels entre 0 et 1
    input_image, input_mask = normalize(input_image, input_mask)

    # === ÉTAPE 4 : CONVERSION DU MASQUE ===
    # tf.cast : Convertit le type de données
    # uint8 : Entier non signé sur 8 bits (valeurs de 0 à 255)
    # Les masques contiennent des classes (0, 1, 2) donc on utilise uint8
    # SUBTRACTION DE 1 POUR QUE LES LABELS SOIENT 0-INDEXÉS (0, 1, 2) au lieu de (1, 2, 3)
    input_mask = tf.cast(input_mask, tf.uint8) - 1

    # Retourne l'image et le masque prétraités
    return input_image, input_mask


# ========== CHARGEMENT DU DATASET ==========

# Affiche un message informatif
print("Chargement du dataset Oxford-IIIT Pet...")

# tfds.load : Charge un dataset depuis TensorFlow Datasets
# 'oxford_iiit_pet:3.*.*' : Nom et version du dataset (chiens et chats)
# with_info=True : Retourne aussi les informations sur le dataset
dataset, info = tfds.load('oxford_iiit_pet:4.*.*', with_info=True)

# === AFFICHAGE DES INFORMATIONS ===

# info.splits : Dictionnaire contenant les différentes parties du dataset
# ['train'] : Partie d'entraînement
# .num_examples : Nombre d'exemples dans cette partie
print(f"Nombre d'exemples d'entraînement: {info.splits['train'].num_examples}")

# ['test'] : Partie de test
print(f"Nombre d'exemples de test: {info.splits['test'].num_examples}")


# ========== PRÉPARATION DU DATASET D'ENTRAÎNEMENT ==========

# dataset['train'] : Sélectionne la partie d'entraînement
# .map() : Applique une fonction à chaque élément du dataset
train_dataset = dataset['train'].map(
    # lambda x : Fonction anonyme (courte fonction sans nom)
    # preprocess(x, augment_data=True) : Applique le prétraitement avec augmentation
    lambda x: preprocess(x, augment_data=True),

    # num_parallel_calls : Nombre de threads pour paralléliser le traitement
    # tf.data.AUTOTUNE : TensorFlow choisit automatiquement le nombre optimal
    # Cela accélère considérablement le chargement des données
    num_parallel_calls=tf.data.AUTOTUNE
)


# ========== PRÉPARATION DU DATASET DE TEST ==========

# Même principe que pour l'entraînement, mais SANS augmentation
# augment_data=False : Pas de transformations aléatoires sur les données de test
train_dataset = dataset['train'].map(
    lambda x: preprocess(x, augment_data=True),
    num_parallel_calls=tf.data.AUTOTUNE
)

test_dataset = dataset['test'].map(
    # augment_data=False : Pas d'augmentation pour le test
    # On veut évaluer sur des données "normales", non modifiées
    lambda x: preprocess(x, augment_data=False),
    num_parallel_calls=tf.data.AUTOTUNE
)


# ========== OPTIMISATION DES DATASETS POUR L'ENTRAÎNEMENT ==========

# === DATASET D'ENTRAÎNEMENT ===
train_dataset = train_dataset.cache() \
                             .shuffle(BUFFER_SIZE) \
                             .batch(BATCH_SIZE) \
                             .prefetch(tf.data.AUTOTUNE)

# Explication détaillée de chaque méthode :

# .cache() : Met en cache le dataset en mémoire après le premier passage
#            Évite de recharger/retraiter les données à chaque epoch
#            Accélère considérablement l'entraînement

# .shuffle(BUFFER_SIZE) : Mélange aléatoirement les données
#                         BUFFER_SIZE=1000 : Mélange 1000 images à la fois
#                         Important pour que le modèle ne voie pas toujours
#                         les mêmes images dans le même ordre

# .batch(BATCH_SIZE) : Regroupe les images en lots de 16
#                      Le modèle traitera 16 images à la fois
#                      Plus efficace que de traiter une image à la fois

# .prefetch(tf.data.AUTOTUNE) : Prépare les données à l'avance
#                               Pendant que le GPU traite un batch,
#                               le CPU prépare le batch suivant
#                               Réduit le temps d'attente


# === DATASET DE TEST ===
test_dataset = test_dataset.batch(BATCH_SIZE) \
                           .prefetch(tf.data.AUTOTUNE)

# Pour le test, on ne fait PAS de cache() ni shuffle()
# .batch(BATCH_SIZE) : Regroupe en lots de 16
# .prefetch() : Prépare les données à l'avance

# Affiche un message de confirmation
print("Datasets préparés avec succès!")
```

```python
# ========== FONCTION DE VISUALISATION DES ÉCHANTILLONS ==========

def display_samples(dataset, num_samples=3):
    """
    Affiche des exemples d'images avec leurs masques de segmentation
    Utile pour vérifier que les données sont correctement chargées

    Args:
        dataset : Dataset TensorFlow contenant les images et masques
        num_samples : Nombre d'échantillons à afficher (défaut: 3)
    """

    # === CRÉATION DE LA FIGURE ===
    # plt.figure : Crée une nouvelle figure matplotlib
    # figsize=(15, 5 * num_samples) : Largeur 15 pouces, hauteur selon le nombre d'échantillons
    # 5 * num_samples : Chaque échantillon prend 5 pouces de hauteur
    plt.figure(figsize=(15, 5 * num_samples))

    # === BOUCLE SUR LES ÉCHANTILLONS ===
    # enumerate : Donne l'index (i) et les données (image, mask)
    # dataset.take(num_samples) : Prend seulement les num_samples premiers éléments
    for i, (image, mask) in enumerate(dataset.take(num_samples)):

        # === EXTRACTION DU PREMIER ÉLÉMENT DU BATCH ===
        # image[0] : Prend la première image du batch
        # .numpy() : Convertit le tenseur TensorFlow en array NumPy
        #            Nécessaire pour l'affichage avec matplotlib
        image = image[0].numpy()
        mask = mask[0].numpy()

        # === AFFICHAGE DE L'IMAGE ORIGINALE ===
        # plt.subplot : Crée une sous-figure dans la grille
        # (num_samples, 3, i * 3 + 1) :
        #   - num_samples lignes (une par échantillon)
        #   - 3 colonnes (image, masque, overlay)
        #   - i * 3 + 1 : Position dans la grille (1ère colonne)
        plt.subplot(num_samples, 3, i * 3 + 1)

        # plt.imshow : Affiche l'image
        # image : Image RGB normalisée (valeurs entre 0 et 1)
        plt.imshow(image)

        # plt.title : Ajoute un titre au-dessus de l'image
        plt.title('Image originale')

        # plt.axis('off') : Masque les axes (pas de coordonnées affichées)
        plt.axis('off')

        # === AFFICHAGE DU MASQUE DE SEGMENTATION ===
        # i * 3 + 2 : Position dans la grille (2ème colonne)
        plt.subplot(num_samples, 3, i * 3 + 2)

        # cmap='gray' : Colormap en niveaux de gris
        # Le masque contient des valeurs 0, 1, 2 (classes)
        plt.imshow(mask, cmap='gray')

        plt.title('Masque de segmentation')
        plt.axis('off')

        # === AFFICHAGE DE L'OVERLAY (SUPERPOSITION) ===
        # i * 3 + 3 : Position dans la grille (3ème colonne)
        plt.subplot(num_samples, 3, i * 3 + 3)

        # Affiche d'abord l'image originale
        plt.imshow(image)

        # Superpose le masque avec transparence
        # alpha=0.5 : Transparence à 50%
        # cmap='jet' : Colormap arc-en-ciel (bleu → rouge)
        #              Permet de mieux voir les différentes classes
        plt.imshow(mask, alpha=0.5, cmap='jet')

        plt.title('Overlay')
        plt.axis('off')

    # === AJUSTEMENT DE LA MISE EN PAGE ===
    # tight_layout() : Ajuste automatiquement l'espacement entre les sous-figures
    #                  Évite que les titres ou images ne se chevauchent
    plt.tight_layout()

    # plt.show() : Affiche la figure à l'écran
    plt.show()


# ========== APPEL DE LA FONCTION ==========

# Affiche un message informatif
print("Exemples du dataset d'entraînement:")

# Appelle la fonction pour afficher 3 exemples
# train_dataset : Dataset préparé précédemment
display_samples(train_dataset)
```

```python
# ========== FONCTION DE CRÉATION DU MODÈLE U-NET ==========

def unet_model(input_shape=(IMG_SIZE, IMG_SIZE, 3), num_classes=3):
    """
    Crée l'architecture U-Net pour la segmentation d'images

    U-Net est une architecture en forme de "U" avec :
    - ENCODER (partie gauche) : Réduit la taille, augmente les features
    - BOTTLENECK (fond du U) : Représentation compressée
    - DECODER (partie droite) : Augmente la taille, génère la segmentation
    - SKIP CONNECTIONS : Relie encoder et decoder pour préserver les détails

    Args:
        input_shape : Forme de l'image d'entrée (hauteur, largeur, canaux)
                      Par défaut : (128, 128, 3) pour des images RGB 128x128
        num_classes : Nombre de classes pour la segmentation
                      3 = fond, animal, contour

    Returns:
        model : Modèle Keras U-Net compilé
    """

    # === DÉFINITION DE L'ENTRÉE ===
    # layers.Input : Définit la forme de l'image d'entrée
    # shape=(IMG_SIZE, IMG_SIZE, 3) : Hauteur, Largeur, Canaux (RGB)
    inputs = layers.Input(shape=input_shape)


    # ============================================================
    # ENCODER (PARTIE DESCENDANTE DU U)
    # Objectif : Extraire des features de plus en plus abstraites
    # Chaque bloc réduit la taille spatiale et augmente les canaux
    # ============================================================

    # === BLOC 1 : 128x128 → 64x64 ===

    # layers.Conv2D : Couche de convolution 2D
    # 64 : Nombre de filtres (features maps)
    # 3 : Taille du kernel (3x3 pixels)
    # activation='relu' : Fonction d'activation ReLU (max(0, x))
    # padding='same' : Ajoute du padding pour garder la même taille
    # kernel_initializer='he_normal' : Initialisation des poids (méthode He)
    # (inputs) : Applique la convolution sur l'entrée
    conv1 = layers.Conv2D(64, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(inputs)

    # Deuxième convolution dans le même bloc
    # Double convolution = pattern classique de U-Net
    conv1 = layers.Conv2D(64, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(conv1)

    # layers.MaxPooling2D : Réduit la taille spatiale
    # pool_size=(2, 2) : Divise la hauteur et largeur par 2
    # 128x128 → 64x64
    pool1 = layers.MaxPooling2D(pool_size=(2, 2))(conv1)


    # === BLOC 2 : 64x64 → 32x32 ===

    # 128 filtres : On double le nombre de features à chaque niveau
    conv2 = layers.Conv2D(128, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(pool1)
    conv2 = layers.Conv2D(128, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(conv2)

    # Pooling : 64x64 → 32x32
    pool2 = layers.MaxPooling2D(pool_size=(2, 2))(conv2)


    # === BLOC 3 : 32x32 → 16x16 ===

    # 256 filtres : Continue à doubler
    conv3 = layers.Conv2D(256, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(pool2)
    conv3 = layers.Conv2D(256, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(conv3)

    # Pooling : 32x32 → 16x16
    pool3 = layers.MaxPooling2D(pool_size=(2, 2))(conv3)


    # === BLOC 4 : 16x16 → 8x8 ===

    # 512 filtres : Continue à doubler
    conv4 = layers.Conv2D(512, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(pool3)
    conv4 = layers.Conv2D(512, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(conv4)

    # layers.Dropout : Désactive aléatoirement 50% des neurones
    # 0.5 : Probabilité de désactivation
    # Aide à éviter le surapprentissage (overfitting)
    drop4 = layers.Dropout(0.5)(conv4)

    # Pooling : 16x16 → 8x8
    pool4 = layers.MaxPooling2D(pool_size=(2, 2))(drop4)


    # ============================================================
    # BOTTLENECK (FOND DU U)
    # Taille minimale : 8x8
    # Features maximales : 1024
    # Représentation la plus compressée de l'image
    # ============================================================

    # 1024 filtres : Maximum de features
    conv5 = layers.Conv2D(1024, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(pool4)
    conv5 = layers.Conv2D(1024, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(conv5)

    # Dropout pour éviter l'overfitting
    drop5 = layers.Dropout(0.5)(conv5)


    # ============================================================
    # DECODER (PARTIE ASCENDANTE DU U)
    # Objectif : Reconstruire l'image segmentée
    # Chaque bloc augmente la taille spatiale et réduit les canaux
    # Les skip connections ajoutent les détails de l'encoder
    # ============================================================

    # === BLOC 6 : 8x8 → 16x16 ===

    # layers.Conv2DTranspose : Convolution transposée (déconvolution)
    # Augmente la taille spatiale (upsampling)
    # 512 : Nombre de filtres
    # 2 : Taille du kernel
    # strides=(2, 2) : Multiplie la taille par 2
    # 8x8 → 16x16
    up6 = layers.Conv2DTranspose(512, 2, strides=(2, 2), padding='same')(drop5)

    # SKIP CONNECTION : Concatène avec conv4 (encoder)
    # layers.concatenate : Fusionne deux tenseurs
    # [up6, drop4] : Liste des tenseurs à fusionner
    # axis=3 : Concatène sur l'axe des canaux (profondeur)
    # Permet de récupérer les détails perdus lors du pooling
    up6 = layers.concatenate([up6, drop4], axis=3)

    # Double convolution après la concaténation
    conv6 = layers.Conv2D(512, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(up6)
    conv6 = layers.Conv2D(512, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(conv6)


    # === BLOC 7 : 16x16 → 32x32 ===

    # Upsampling : 16x16 → 32x32
    up7 = layers.Conv2DTranspose(256, 2, strides=(2, 2), padding='same')(conv6)

    # Skip connection avec conv3
    up7 = layers.concatenate([up7, conv3], axis=3)

    # Double convolution
    conv7 = layers.Conv2D(256, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(up7)
    conv7 = layers.Conv2D(256, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(conv7)


    # === BLOC 8 : 32x32 → 64x64 ===

    # Upsampling : 32x32 → 64x64
    up8 = layers.Conv2DTranspose(128, 2, strides=(2, 2), padding='same')(conv7)

    # Skip connection avec conv2
    up8 = layers.concatenate([up8, conv2], axis=3)

    # Double convolution
    conv8 = layers.Conv2D(128, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(up8)
    conv8 = layers.Conv2D(128, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(conv8)


    # === BLOC 9 : 64x64 → 128x128 ===

    # Upsampling final : 64x64 → 128x128
    # Retour à la taille originale de l'image
    up9 = layers.Conv2DTranspose(64, 2, strides=(2, 2), padding='same')(conv8)

    # Skip connection avec conv1
    up9 = layers.concatenate([up9, conv1], axis=3)

    # Double convolution finale
    conv9 = layers.Conv2D(64, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(up9)
    conv9 = layers.Conv2D(64, 3, activation='relu', padding='same',
                          kernel_initializer='he_normal')(conv9)


    # ============================================================
    # COUCHE DE SORTIE
    # Génère la segmentation finale
    # ============================================================

    # Convolution 1x1 pour réduire à num_classes canaux
    # layers.Conv2D(num_classes, 1, ...) :
    #   - num_classes (3) : Un canal par classe (fond, animal, contour)
    #   - 1 : Kernel 1x1 (convolution ponctuelle)
    #   - activation='softmax' : Convertit en probabilités
    #     Chaque pixel aura 3 valeurs entre 0 et 1 qui somment à 1
    #     Représente la probabilité d'appartenir à chaque classe
    outputs = layers.Conv2D(num_classes, 1, activation='softmax')(conv9)


    # === CRÉATION DU MODÈLE ===
    # models.Model : Crée un modèle Keras
    # inputs=inputs : Définit l'entrée du modèle
    # outputs=outputs : Définit la sortie du modèle
    model = models.Model(inputs=inputs, outputs=outputs)

    # Retourne le modèle construit
    return model


# ========== CRÉATION DU MODÈLE ==========

# Appelle la fonction pour créer le modèle U-Net
# input_shape=(IMG_SIZE, IMG_SIZE, 3) : Images 128x128 RGB
# num_classes=3 : 3 classes (fond, animal, contour)
model = unet_model(input_shape=(IMG_SIZE, IMG_SIZE, 3), num_classes=3)


# ========== AFFICHAGE DU RÉSUMÉ ==========

# model.summary() : Affiche l'architecture complète du modèle
# Montre :
#   - Toutes les couches
#   - La forme de sortie de chaque couche
#   - Le nombre de paramètres par couche
#   - Le nombre total de paramètres
model.summary()


# === COMPTAGE DES PARAMÈTRES ===
# model.count_params() : Compte le nombre total de paramètres
# Ce sont les poids que le modèle va apprendre
total_params = model.count_params()

# Affichage avec formatage
# {:,} : Ajoute des virgules pour séparer les milliers
# Exemple : 31043971 devient 31,043,971
print(f"\nNombre total de paramètres: {total_params:,}")
```

```python
# ========== DÉFINITION DES FONCTIONS DE PERTE ==========

# === DICE LOSS ===
def dice_loss(y_true, y_pred, smooth=1e-6):
    """
    Calcule le Dice Loss (coefficient de Dice)

    Le Dice Loss mesure la similarité entre deux ensembles
    Particulièrement utile pour les classes déséquilibrées
    (quand il y a beaucoup plus de pixels de fond que d'objets)

    Formule : Dice = 2 * |A ∩ B| / (|A| + |B|)
    où A = pixels prédits, B = pixels réels

    Args:
        y_true : Masque réel (ground truth)
        y_pred : Masque prédit par le modèle
        smooth : Valeur pour éviter la division par zéro (1e-6 = 0.000001)

    Returns:
        dice_loss : Valeur de perte (plus bas = mieux)
    """

    # tf.keras.backend.flatten : Transforme le tenseur en vecteur 1D
    # Exemple : [[1, 2], [3, 4]] → [1, 2, 3, 4]
    # Facilite le calcul de l'intersection
    y_true_f = tf.keras.backend.flatten(y_true)
    y_pred_f = tf.keras.backend.flatten(y_pred)

    # Calcul de l'intersection
    # tf.keras.backend.sum : Somme tous les éléments
    # y_true_f * y_pred_f : Multiplication élément par élément
    # Pixels où prédiction ET vérité = 1
    intersection = tf.keras.backend.sum(y_true_f * y_pred_f)

    # Formule du Dice Loss
    # 1 - Dice : Convertit le coefficient de Dice en perte
    # Dice proche de 1 → Dice Loss proche de 0 (bon)
    # Dice proche de 0 → Dice Loss proche de 1 (mauvais)
    # smooth : Évite la division par zéro si intersection = 0
    dice_coefficient = (2. * intersection + smooth) / \
                      (tf.keras.backend.sum(y_true_f) +
                       tf.keras.backend.sum(y_pred_f) + smooth)

    return 1 - dice_coefficient


# === FONCTION DE PERTE COMBINÉE ===
def combined_loss(y_true, y_pred):
    """
    Combine Dice Loss et Cross-Entropy Loss

    Cross-Entropy : Mesure la différence entre distributions de probabilités
    Dice Loss : Mesure la similarité géométrique

    Combiner les deux donne souvent de meilleurs résultats

    Args:
        y_true : Masque réel
        y_pred : Masque prédit

    Returns:
        total_loss : Somme des deux pertes
    """

    # sparse_categorical_crossentropy : Fonction de perte pour classification
    # "sparse" : Les labels sont des entiers (0, 1, 2) pas one-hot encoded
    # "categorical" : Pour plusieurs classes
    # Mesure l'écart entre prédiction et vérité
    ce_loss = tf.keras.losses.sparse_categorical_crossentropy(y_true, y_pred)

    # Calcul du Dice Loss
    dice = dice_loss(y_true, y_pred)

    # Retourne la somme des deux pertes
    # On peut aussi utiliser une moyenne pondérée : alpha*ce_loss + beta*dice
    return ce_loss + dice


# ========== DÉFINITION DES MÉTRIQUES ==========

# === MÉTRIQUE IoU (Intersection over Union) ===
class MeanIoU(tf.keras.metrics.MeanIoU):
    """
    Métrique IoU moyenne pour évaluer la segmentation

    IoU = Aire de l'intersection / Aire de l'union

    Hérite de tf.keras.metrics.MeanIoU et adapte pour notre cas
    """

    def update_state(self, y_true, y_pred, sample_weight=None):
        """
        Met à jour l'état de la métrique avec de nouvelles prédictions

        Args:
            y_true : Masques réels
            y_pred : Masques prédits (probabilités)
            sample_weight : Poids des échantillons (optionnel)

        Returns:
            Résultat de la mise à jour de l'état parent
        """

        # tf.squeeze(y_true, axis=-1) : Supprime la dimension inutile
        # Exemple : (16, 128, 128, 1) → (16, 128, 128)

        # tf.argmax(y_pred, axis=-1) : Prend la classe avec la plus haute probabilité
        # y_pred contient 3 valeurs par pixel (probabilités pour chaque classe)
        # argmax retourne l'index de la classe la plus probable
        # Exemple : [0.1, 0.7, 0.2] → 1 (classe 1 a la plus haute probabilité)

        # Appelle la méthode update_state de la classe parente
        return super().update_state(
            tf.squeeze(y_true, axis=-1),  # Masque réel nettoyé
            tf.argmax(y_pred, axis=-1),   # Prédiction convertie en classe
            sample_weight                  # Poids (si fourni)
        )


# ========== COMPILATION DU MODÈLE ==========

# model.compile : Configure le modèle pour l'entraînement
model.compile(

    # === OPTIMISEUR ===
    # optimizer : Algorithme qui ajuste les poids du réseau
    # Adam : Adaptive Moment Estimation
    #   - Algorithme d'optimisation très populaire
    #   - Combine les avantages de RMSprop et SGD with momentum
    #   - Adapte le learning rate pour chaque paramètre
    # learning_rate=1e-4 : Taux d'apprentissage (0.0001)
    #   - Petit learning rate = apprentissage lent mais stable
    #   - Grand learning rate = apprentissage rapide mais instable
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4),

    # === FONCTION DE PERTE ===
    # loss : Mesure l'écart entre prédiction et vérité
    # 'sparse_categorical_crossentropy' : Pour classification multi-classes
    # Alternative : Utiliser combined_loss pour meilleurs résultats
    loss='sparse_categorical_crossentropy',
    # loss=combined_loss,  # Décommentez pour utiliser la perte combinée

    # === MÉTRIQUES ===
    # metrics : Liste des métriques à calculer pendant l'entraînement
    # Ne sont PAS utilisées pour l'entraînement, seulement pour monitoring
    metrics=[
        'accuracy',              # Précision globale (% pixels corrects)
        MeanIoU(num_classes=3)   # IoU moyen sur les 3 classes
    ]
)


# ========== DÉFINITION DES CALLBACKS ==========
# Callbacks : Fonctions appelées automatiquement pendant l'entraînement

# === CHECKPOINT : SAUVEGARDE DU MEILLEUR MODÈLE ===
# ModelCheckpoint : Sauvegarde le modèle à intervalles réguliers
checkpoint_callback = ModelCheckpoint(
    # Nom du fichier où sauvegarder le modèle
    'unet_best_model.h5',

    # monitor='val_loss' : Métrique à surveiller
    # Le modèle sera sauvegardé quand val_loss s'améliore
    monitor='val_loss',

    # save_best_only=True : Sauvegarde seulement le meilleur modèle
    # Évite de sauvegarder à chaque epoch
    save_best_only=True,

    # mode='min' : Plus bas = mieux (pour la loss)
    # Alternative : mode='max' pour accuracy
    mode='min',

    # verbose=1 : Affiche un message lors de la sauvegarde
    verbose=1
)


# === EARLY STOPPING : ARRÊT ANTICIPÉ ===
# EarlyStopping : Arrête l'entraînement si le modèle ne s'améliore plus
early_stopping = EarlyStopping(
    # monitor='val_loss' : Métrique à surveiller
    monitor='val_loss',

    # patience=10 : Nombre d'epochs sans amélioration avant d'arrêter
    # Si val_loss ne diminue pas pendant 10 epochs → arrêt
    patience=10,

    # restore_best_weights=True : Restaure les poids du meilleur epoch
    # Évite de garder les poids de la dernière epoch (potentiellement pire)
    restore_best_weights=True,

    # verbose=1 : Affiche un message lors de l'arrêt
    verbose=1
)


# === REDUCE LEARNING RATE : RÉDUCTION DU TAUX D'APPRENTISSAGE ===
# ReduceLROnPlateau : Réduit le learning rate si le modèle stagne
reduce_lr = tf.keras.callbacks.ReduceLROnPlateau(
    # monitor='val_loss' : Métrique à surveiller
    monitor='val_loss',

    # factor=0.5 : Multiplie le learning rate par 0.5
    # Exemple : 0.0001 → 0.00005
    factor=0.5,

    # patience=5 : Réduit le learning rate après 5 epochs sans amélioration
    patience=5,

    # min_lr=1e-7 : Learning rate minimum (0.0000001)
    # En dessous, ne réduit plus
    min_lr=1e-7,

    # verbose=1 : Affiche un message lors de la réduction
    verbose=1
)


# ========== LANCEMENT DE L'ENTRAÎNEMENT ==========

# Nombre d'epochs (passages complets sur les données)
# 25 epochs est généralement suffisant pour U-Net
EPOCHS = 25

# Message informatif
print("Début de l'entraînement...")

# model.fit : Lance l'entraînement du modèle
history = model.fit(

    # === DONNÉES D'ENTRAÎNEMENT ===
    # train_dataset : Dataset préparé avec augmentation
    train_dataset,

    # === NOMBRE D'EPOCHS ===
    # epochs : Nombre de fois où le modèle verra tout le dataset
    epochs=EPOCHS,

    # === DONNÉES DE VALIDATION ===
    # validation_data : Dataset pour évaluer pendant l'entraînement
    # Utilisé pour calculer val_loss et val_accuracy
    # IMPORTANT : Pas utilisé pour l'entraînement, seulement pour évaluation
    validation_data=test_dataset,

    # === CALLBACKS ===
    # callbacks : Liste des fonctions à appeler pendant l'entraînement
    callbacks=[
        checkpoint_callback,  # Sauvegarde le meilleur modèle
        early_stopping,       # Arrête si pas d'amélioration
        reduce_lr            # Réduit le learning rate si stagnation
    ],

    # === VERBOSITY ===
    # verbose=1 : Affiche une barre de progression pour chaque epoch
    # verbose=0 : Mode silencieux
    # verbose=2 : Affiche une ligne par epoch
    verbose=1
)

# Message de fin
print("\nEntraînement terminé!")

# history : Objet contenant l'historique de l'entraînement
# history.history : Dictionnaire avec les métriques par epoch
#   - 'loss' : Loss d'entraînement
#   - 'val_loss' : Loss de validation
#   - 'accuracy' : Accuracy d'entraînement
#   - 'val_accuracy' : Accuracy de validation
#   - etc.
```

```python
# ========== FONCTION DE CALCUL DES MÉTRIQUES ==========

def calculate_metrics(y_true, y_pred):
    """
    Calcule les métriques de segmentation : IoU, Précision, Rappel, F1-Score

    Ces métriques permettent d'évaluer la qualité de la segmentation

    Args:
        y_true : Masque réel (valeurs binaires 0 ou 1)
        y_pred : Masque prédit (valeurs binaires 0 ou 1)

    Returns:
        dict : Dictionnaire contenant les 4 métriques
    """

    # === APLATISSEMENT DES TENSEURS ===
    # .flatten() : Transforme le tableau 2D en vecteur 1D
    # Exemple : [[1, 0], [1, 1]] → [1, 0, 1, 1]
    # Facilite les calculs
    y_true_flat = y_true.flatten()
    y_pred_flat = y_pred.flatten()

    # === CALCUL DE L'INTERSECTION ===
    # np.sum() : Somme tous les éléments
    # y_true_flat * y_pred_flat : Multiplication élément par élément
    # Résultat = 1 seulement si les deux valent 1
    # Compte les pixels correctement prédits comme objet
    intersection = np.sum(y_true_flat * y_pred_flat)

    # === CALCUL DE L'UNION ===
    # Union = A + B - Intersection
    # Pixels qui sont dans A OU dans B (ou les deux)
    union = np.sum(y_true_flat) + np.sum(y_pred_flat) - intersection

    # === IoU (Intersection over Union) ===
    # IoU = Intersection / Union
    # Mesure le chevauchement entre prédiction et vérité
    # Valeur entre 0 (aucun chevauchement) et 1 (parfait)
    # + 1e-7 : Évite la division par zéro
    iou = intersection / (union + 1e-7)

    # === PRÉCISION (Precision) ===
    # Précision = Vrais Positifs / (Vrais Positifs + Faux Positifs)
    # Parmi les pixels prédits comme objet, combien sont vraiment objet?
    # Mesure la qualité des prédictions positives
    precision = intersection / (np.sum(y_pred_flat) + 1e-7)

    # === RAPPEL (Recall) ===
    # Rappel = Vrais Positifs / (Vrais Positifs + Faux Négatifs)
    # Parmi les vrais pixels d'objet, combien ont été détectés?
    # Mesure la capacité à détecter tous les objets
    recall = intersection / (np.sum(y_true_flat) + 1e-7)

    # === F1-SCORE ===
    # F1 = 2 * (Précision * Rappel) / (Précision + Rappel)
    # Moyenne harmonique de la précision et du rappel
    # Équilibre entre précision et rappel
    # Valeur entre 0 (mauvais) et 1 (parfait)
    f1 = 2 * (precision * recall) / (precision + recall + 1e-7)

    # === RETOUR DES RÉSULTATS ===
    # Retourne un dictionnaire avec les 4 métriques
    return {
        'IoU': iou,              # Intersection over Union
        'Precision': precision,  # Précision
        'Recall': recall,        # Rappel
        'F1-Score': f1          # F1-Score
    }


# ========== ÉVALUATION SUR LE DATASET DE TEST ==========

# Message informatif
print("Évaluation du modèle sur le dataset de test...")

# model.evaluate : Évalue le modèle sur un dataset
# test_dataset : Dataset de test (jamais vu pendant l'entraînement)
# Retourne : [loss, metric1, metric2, ...]
# Dans notre cas : [loss, accuracy, mean_iou]
test_loss, test_accuracy, test_iou = model.evaluate(test_dataset)


# ========== AFFICHAGE DES RÉSULTATS ==========

# Affichage formaté avec des séparateurs
print(f"\n{'='*50}")
print(f"Résultats sur le dataset de test:")
print(f"{'='*50}")

# f"..." : f-string pour formater les chaînes
# {test_loss:.4f} : Affiche test_loss avec 4 décimales
# .4f : Format à 4 chiffres après la virgule
print(f"Loss: {test_loss:.4f}")
print(f"Accuracy: {test_accuracy:.4f}")
print(f"Mean IoU: {test_iou:.4f}")

print(f"{'='*50}")

# Interprétation des résultats :
# Loss proche de 0 = bon modèle
# Accuracy proche de 1 (100%) = bon modèle
# Mean IoU > 0.7 = très bon modèle
# Mean IoU > 0.5 = modèle acceptable
# Mean IoU < 0.5 = modèle faible
```

```python
# ========== FONCTION D'AFFICHAGE DE L'HISTORIQUE D'ENTRAÎNEMENT ==========

def plot_training_history(history):
    """
    Affiche les courbes d'évolution des métriques pendant l'entraînement

    Permet de visualiser :
    - Si le modèle apprend (loss diminue)
    - Si le modèle sur-apprend (val_loss augmente)
    - L'évolution de l'accuracy et de l'IoU

    Args:
        history : Objet history retourné par model.fit()
    """

    # === CRÉATION DE LA FIGURE ===
    # fig, axes : Crée une figure avec plusieurs sous-graphiques
    # 1, 3 : 1 ligne, 3 colonnes
    # figsize=(18, 5) : Largeur 18 pouces, hauteur 5 pouces
    fig, axes = plt.subplots(1, 3, figsize=(18, 5))

    # === GRAPHIQUE 1 : LOSS ===
    # axes[0] : Premier sous-graphique (index 0)

    # .plot() : Trace une courbe
    # history.history['loss'] : Liste des valeurs de loss par epoch
    # label='Train Loss' : Légende pour cette courbe
    axes[0].plot(history.history['loss'], label='Train Loss')

    # Trace la loss de validation
    # 'val_loss' : Loss calculée sur le dataset de validation
    axes[0].plot(history.history['val_loss'], label='Val Loss')

    # .set_title() : Définit le titre du graphique
    axes[0].set_title('Évolution de la Loss')

    # .set_xlabel() : Définit le label de l'axe X
    axes[0].set_xlabel('Epoch')

    # .set_ylabel() : Définit le label de l'axe Y
    axes[0].set_ylabel('Loss')

    # .legend() : Affiche la légende (Train Loss, Val Loss)
    axes[0].legend()

    # .grid(True) : Affiche une grille en arrière-plan
    axes[0].grid(True)

    # === GRAPHIQUE 2 : ACCURACY ===
    # axes[1] : Deuxième sous-graphique

    # Trace l'accuracy d'entraînement
    axes[1].plot(history.history['accuracy'], label='Train Accuracy')

    # Trace l'accuracy de validation
    axes[1].plot(history.history['val_accuracy'], label='Val Accuracy')

    axes[1].set_title('Évolution de l\'Accuracy')
    axes[1].set_xlabel('Epoch')
    axes[1].set_ylabel('Accuracy')
    axes[1].legend()
    axes[1].grid(True)

    # === GRAPHIQUE 3 : IoU ===
    # axes[2] : Troisième sous-graphique

    # 'mean_io_u' : Nom de la métrique IoU dans history
    # (le nom peut varier selon la version de TensorFlow)
    axes[2].plot(history.history['mean_io_u_2'], label='Train IoU')
    axes[2].plot(history.history['val_mean_io_u_2'], label='Val IoU')

    axes[2].set_title('Évolution de l\'IoU')
    axes[2].set_xlabel('Epoch')
    axes[2].set_ylabel('IoU')
    axes[2].legend()
    axes[2].grid(True)

    # === AJUSTEMENT DE LA MISE EN PAGE ===
    # tight_layout() : Ajuste automatiquement l'espacement
    # Évite que les graphiques se chevauchent
    plt.tight_layout()

    # Affiche la figure
    plt.show()


# ========== APPEL DE LA FONCTION ==========
# Affiche les courbes d'entraînement
plot_training_history(history)


# ========== FONCTION DE VISUALISATION DES PRÉDICTIONS ==========

def display_predictions(dataset, model, num_samples=5):
    """
    Affiche les prédictions du modèle sur des exemples

    Pour chaque exemple, affiche :
    - L'image originale
    - Le masque réel (ground truth)
    - La prédiction du modèle
    - L'overlay (superposition prédiction sur image)

    Args:
        dataset : Dataset contenant les images et masques
        model : Modèle entraîné
        num_samples : Nombre d'exemples à afficher
    """

    # === CRÉATION DE LA FIGURE ===
    # 5 * num_samples : Hauteur selon le nombre d'échantillons
    # 4 colonnes : Image, Masque réel, Prédiction, Overlay
    plt.figure(figsize=(15, 5 * num_samples))

    # === BOUCLE SUR LES ÉCHANTILLONS ===
    # enumerate : Donne l'index (i) et les données (image, mask)
    # dataset.take(num_samples) : Prend seulement num_samples exemples
    for i, (image, mask) in enumerate(dataset.take(num_samples)):

        # === PRÉDICTION ===
        # model.predict() : Fait une prédiction sur l'image
        # image : Batch d'images (shape: [batch_size, 128, 128, 3])
        # verbose=0 : Pas de barre de progression
        # Retourne : Probabilités pour chaque classe [batch_size, 128, 128, 3]
        pred_mask = model.predict(image, verbose=0)

        # tf.argmax() : Prend l'index de la valeur maximale
        # axis=-1 : Sur le dernier axe (les 3 classes)
        # Convertit les probabilités en classe prédite
        # Exemple : [0.1, 0.7, 0.2] → 1 (classe 1)
        pred_mask = tf.argmax(pred_mask, axis=-1)

        # [0] : Prend le premier élément du batch
        # .numpy() : Convertit en array NumPy
        pred_mask = pred_mask[0].numpy()

        # === EXTRACTION DES DONNÉES ===
        # Prend la première image du batch et convertit en NumPy
        image = image[0].numpy()
        mask = mask[0].numpy()

        # === AFFICHAGE DE L'IMAGE ORIGINALE ===
        # (num_samples, 4, i * 4 + 1) : Position dans la grille
        #   - num_samples lignes
        #   - 4 colonnes
        #   - i * 4 + 1 : Première colonne de la ligne i
        plt.subplot(num_samples, 4, i * 4 + 1)
        plt.imshow(image)
        plt.title('Image')
        plt.axis('off')

        # === AFFICHAGE DU MASQUE RÉEL ===
        # i * 4 + 2 : Deuxième colonne
        plt.subplot(num_samples, 4, i * 4 + 2)

        # cmap='gray' : Colormap en niveaux de gris
        plt.imshow(mask, cmap='gray')
        plt.title('Masque Réel')
        plt.axis('off')

        # === AFFICHAGE DE LA PRÉDICTION ===
        # i * 4 + 3 : Troisième colonne
        plt.subplot(num_samples, 4, i * 4 + 3)
        plt.imshow(pred_mask, cmap='gray')
        plt.title('Prédiction')
        plt.axis('off')

        # === AFFICHAGE DE L'OVERLAY ===
        # i * 4 + 4 : Quatrième colonne
        plt.subplot(num_samples, 4, i * 4 + 4)

        # Affiche d'abord l'image
        plt.imshow(image)

        # Superpose la prédiction avec transparence
        # alpha=0.5 : Transparence à 50%
        # cmap='jet' : Colormap arc-en-ciel
        plt.imshow(pred_mask, alpha=0.5, cmap='jet')
        plt.title('Overlay')
        plt.axis('off')

    # Ajuste la mise en page
    plt.tight_layout()
    plt.show()


# ========== APPEL DE LA FONCTION ==========
print("Visualisation des prédictions:")
display_predictions(test_dataset, model, num_samples=5)


# ========== FONCTION DE VISUALISATION DES ERREURS ==========

def display_error_heatmap(dataset, model, num_samples=3):
    """
    Affiche une heatmap (carte de chaleur) des erreurs de prédiction

    La heatmap montre où le modèle fait des erreurs :
    - Zones sombres : Peu d'erreurs
    - Zones claires/chaudes : Beaucoup d'erreurs

    Args:
        dataset : Dataset contenant les images et masques
        model : Modèle entraîné
        num_samples : Nombre d'exemples à afficher
    """

    # === CRÉATION DE LA FIGURE ===
    # 3 colonnes : Image, Prédiction, Heatmap des erreurs
    plt.figure(figsize=(15, 5 * num_samples))

    # === BOUCLE SUR LES ÉCHANTILLONS ===
    for i, (image, mask) in enumerate(dataset.take(num_samples)):

        # === PRÉDICTION ===
        # Même processus que dans display_predictions
        pred_mask = model.predict(image, verbose=0)
        pred_mask = tf.argmax(pred_mask, axis=-1)
        pred_mask = pred_mask[0].numpy()

        # === EXTRACTION DES DONNÉES ===
        image = image[0].numpy()

        # .squeeze() : Supprime les dimensions de taille 1
        # Exemple : (128, 128, 1) → (128, 128)
        mask = mask[0].numpy().squeeze()

        # === CALCUL DES ERREURS ===
        # np.abs() : Valeur absolue
        # mask - pred_mask : Différence entre vérité et prédiction
        # Résultat = 0 : Prédiction correcte
        # Résultat ≠ 0 : Erreur de prédiction
        error_map = np.abs(mask - pred_mask)

        # === AFFICHAGE DE L'IMAGE ===
        # i * 3 + 1 : Première colonne
        plt.subplot(num_samples, 3, i * 3 + 1)
        plt.imshow(image)
        plt.title('Image')
        plt.axis('off')

        # === AFFICHAGE DE LA PRÉDICTION ===
        # i * 3 + 2 : Deuxième colonne
        plt.subplot(num_samples, 3, i * 3 + 2)
        plt.imshow(pred_mask, cmap='gray')
        plt.title('Prédiction')
        plt.axis('off')

        # === AFFICHAGE DE LA HEATMAP DES ERREURS ===
        # i * 3 + 3 : Troisième colonne
        plt.subplot(num_samples, 3, i * 3 + 3)

        # cmap='hot' : Colormap "chaleur"
        #   - Noir : Pas d'erreur (error = 0)
        #   - Rouge/Jaune : Erreurs (error > 0)
        # Plus la couleur est chaude, plus l'erreur est importante
        plt.imshow(error_map, cmap='hot')

        # plt.colorbar() : Ajoute une barre de couleur
        # Indique la correspondance entre couleurs et valeurs
        plt.colorbar()

        plt.title('Heatmap des Erreurs')
        plt.axis('off')

    # Ajuste la mise en page
    plt.tight_layout()
    plt.show()


# ========== APPEL DE LA FONCTION ==========
print("Visualisation des erreurs:")
display_error_heatmap(test_dataset, model, num_samples=3)
```

```python
# ========== FONCTION DE PRÉDICTION SUR UNE IMAGE UNIQUE ==========

def predict_single_image(model, image_path):
    """
    Fait une prédiction sur une image unique

    Permet d'utiliser le modèle entraîné sur n'importe quelle image

    Args:
        model : Modèle entraîné
        image_path : Chemin vers l'image (ex: 'mon_chat.jpg')

    Returns:
        pred_mask : Masque de segmentation prédit
    """

    # === CHARGEMENT DE L'IMAGE ===
    # cv2.imread() : Lit une image depuis un fichier
    # Retourne un array NumPy avec les pixels
    # Format : BGR (Blue, Green, Red) - ordre inversé !
    img = cv2.imread(image_path)

    # === CONVERSION BGR → RGB ===
    # cv2.cvtColor() : Convertit l'espace de couleur
    # cv2.COLOR_BGR2RGB : De BGR vers RGB
    # Nécessaire car OpenCV utilise BGR mais le modèle attend RGB
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

    # === REDIMENSIONNEMENT ===
    # cv2.resize() : Redimensionne l'image
    # (IMG_SIZE, IMG_SIZE) : Nouvelle taille (128x128)
    # Le modèle a été entraîné sur 128x128, donc on doit redimensionner
    img_resized = cv2.resize(img, (IMG_SIZE, IMG_SIZE))

    # === NORMALISATION ===
    # / 255.0 : Normalise les pixels de [0, 255] vers [0, 1]
    # Le modèle a été entraîné sur des données normalisées
    img_normalized = img_resized / 255.0

    # === AJOUT D'UNE DIMENSION BATCH ===
    # np.expand_dims() : Ajoute une dimension
    # axis=0 : Ajoute au début
    # (128, 128, 3) → (1, 128, 128, 3)
    # Le modèle attend un batch, même si on n'a qu'une image
    img_batch = np.expand_dims(img_normalized, axis=0)

    # === PRÉDICTION ===
    # model.predict() : Fait une prédiction
    # img_batch : Image préparée
    # verbose=0 : Pas de barre de progression
    # Retourne : Probabilités pour chaque classe
    pred = model.predict(img_batch, verbose=0)

    # === CONVERSION EN CLASSE ===
    # pred[0] : Prend la première (et unique) image du batch
    # tf.argmax(..., axis=-1) : Prend la classe avec la plus haute probabilité
    # .numpy() : Convertit en array NumPy
    pred_mask = tf.argmax(pred[0], axis=-1).numpy()

    # === VISUALISATION ===
    # Crée une figure avec 3 sous-graphiques
    plt.figure(figsize=(15, 5))

    # === IMAGE ORIGINALE ===
    # 1, 3, 1 : 1 ligne, 3 colonnes, position 1
    plt.subplot(1, 3, 1)
    plt.imshow(img_resized)
    plt.title('Image Originale')
    plt.axis('off')

    # === MASQUE PRÉDIT ===
    # Position 2
    plt.subplot(1, 3, 2)
    plt.imshow(pred_mask, cmap='gray')
    plt.title('Masque Prédit')
    plt.axis('off')

    # === OVERLAY ===
    # Position 3
    plt.subplot(1, 3, 3)
    plt.imshow(img_resized)
    plt.imshow(pred_mask, alpha=0.5, cmap='jet')
    plt.title('Overlay')
    plt.axis('off')

    # Ajuste la mise en page
    plt.tight_layout()

    # Affiche la figure
    plt.show()

    # Retourne le masque prédit
    return pred_mask


# ========== UTILISATION DE LA FONCTION ==========
# Décommentez et modifiez pour utiliser sur votre propre image

# Exemple d'utilisation :
# mask = predict_single_image(model, 'chemin/vers/votre/image.jpg')

# Exemples de chemins valides :
# - Windows : 'C:/Users/VotreNom/Images/chat.jpg'
# - Mac/Linux : '/home/VotreNom/Images/chat.jpg'
# - Relatif : 'images/mon_animal.jpg' (si le fichier est dans un dossier images/)
```
